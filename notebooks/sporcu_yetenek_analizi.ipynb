````python
# Gerekli kÃ¼tÃ¼phaneleri yÃ¼kle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# TÃ¼rkÃ§e karakter desteÄŸi
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['axes.unicode_minus'] = False

print("âœ… KÃ¼tÃ¼phaneler baÅŸarÄ±yla yÃ¼klendi!")
````

````python
# Veri setini yÃ¼kle ve ilk analiz
data = pd.read_csv("../data/sporcu_dataset.csv")

print("=== VERÄ° SETÄ° GENEL BÄ°LGÄ°LERÄ° ===")
print(f"ğŸ“Š Veri boyutu: {data.shape}")
print(f"ğŸ“Š Ã–zellik sayÄ±sÄ±: {data.shape[1]}")
print(f"ğŸ“Š KiÅŸi sayÄ±sÄ±: {data.shape[0]}")
print(f"ğŸ“Š Eksik deÄŸer sayÄ±sÄ±: {data.isnull().sum().sum()}")

print("\n=== SÃœTUN BÄ°LGÄ°LERÄ° ===")
print(f"ğŸ“‹ SÃ¼tun isimleri: {list(data.columns)}")

print("\n=== HEDEF DEÄÄ°ÅKEN ANALÄ°ZÄ° ===")
print("ğŸ† Spor daÄŸÄ±lÄ±mÄ±:")
print(data['tavsiye_edilen_spor'].value_counts())

# Ä°lk 5 satÄ±rÄ± gÃ¶ster
print("\n=== Ä°LK 5 KAYIT ===")
data.head()
````

````python
# Veri gÃ¶rselleÅŸtirme
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Temel Veri Analizi', fontsize=16, fontweight='bold')

# YaÅŸ daÄŸÄ±lÄ±mÄ±
axes[0, 0].hist(data['yas'], bins=15, color='skyblue', alpha=0.7, edgecolor='black')
axes[0, 0].set_title('YaÅŸ DaÄŸÄ±lÄ±mÄ±')
axes[0, 0].set_xlabel('YaÅŸ')
axes[0, 0].set_ylabel('KiÅŸi SayÄ±sÄ±')
axes[0, 0].grid(True, alpha=0.3)

# Spor daÄŸÄ±lÄ±mÄ±
sport_counts = data['tavsiye_edilen_spor'].value_counts()
axes[0, 1].pie(sport_counts.values, labels=sport_counts.index, autopct='%1.1f%%')
axes[0, 1].set_title('Spor DaÄŸÄ±lÄ±mÄ±')

# BMI daÄŸÄ±lÄ±mÄ±
axes[1, 0].hist(data['bmi'], bins=15, color='lightgreen', alpha=0.7, edgecolor='black')
axes[1, 0].set_title('BMI DaÄŸÄ±lÄ±mÄ±')
axes[1, 0].set_xlabel('BMI')
axes[1, 0].set_ylabel('KiÅŸi SayÄ±sÄ±')
axes[1, 0].grid(True, alpha=0.3)

# Performans korelasyonu
perf_features = ['hiz', 'kuvvet', 'dayaniklilik', 'esneklik', 'koordinasyon']
corr_matrix = data[perf_features].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, ax=axes[1, 1])
axes[1, 1].set_title('Performans Korelasyonu')

plt.tight_layout()
plt.show()

print("âœ… Veri gÃ¶rselleÅŸtirme tamamlandÄ±!")
````
